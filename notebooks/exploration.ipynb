# Exploration Notebook

Use this notebook to load simulation logs, visualize fatigue trajectories, and prototype new reward structures.

```python
import json
from pathlib import Path
import matplotlib.pyplot as plt

log_files = sorted(Path("../results/logs").glob("run_*.json"))
if not log_files:
    raise FileNotFoundError("No logs found. Run `python -m src.main` first.")

payload = json.loads(log_files[-1].read_text())
print(payload["summary"])

plt.plot(payload["trajectories"][0]["fatigue"], label="fatigue")
plt.plot(payload["trajectories"][0]["rewards"], label="reward")
plt.legend()
plt.show()
```
